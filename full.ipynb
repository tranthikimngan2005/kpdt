```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
```

```python
G = nx.Graph()

with open("data.txt", "r") as f:
    for line in f:
        u, v = map(int, line.split())
        G.add_edge(u, v)

print("Load graph thành công!")
```

```python
plt.figure(figsize=(8,6))
nx.draw(G, with_labels=True, node_size=300)
plt.title("Graph Visualization")
plt.show()
```

```python
print("Số node:", G.number_of_nodes())
print("Số cạnh:", G.number_of_edges())
print("Density:", nx.density(G))
```

```python
#Degree của các node
degree = dict(G.degree())
print("Degree của từng node:")
print(degree)
```

```python
#degree centrality
degree_centrality = nx.degree_centrality(G)
print("Degree Centrality:")
print(degree_centrality)
```

```python
closeness = nx.closeness_centrality(G)
print("Closeness Centrality:")
print(closeness)
```

```python
betweenness = nx.betweenness_centrality(G)
print("Betweenness Centrality:")
print(betweenness)
```

```python
df = pd.DataFrame({
    "node": list(degree_centrality.keys()),
    "value": list(degree_centrality.values())
})

top_nodes = df.sort_values("value", ascending=False).head(5)
print("Top 5 node quan trọng nhất:")
print(top_nodes)
```

```python
is_connected = nx.is_connected(G)
print("Graph có liên thông không?", is_connected)
```

```python
components = list(nx.connected_components(G))
print("Các thành phần liên thông:")
print(components)
```

```python
dfs_tree = nx.dfs_tree(G, source=1)
print("DFS từ node 1:")
print(list(dfs_tree))

nx.draw(dfs_tree, with_labels=True)
plt.title("DFS Tree")
plt.show()
```

```python
bfs_tree = nx.bfs_tree(G, source=1)
print("BFS từ node 1:")
print(list(bfs_tree))

nx.draw(bfs_tree, with_labels=True)
plt.title("BFS Tree")
plt.show()
```

```python
path = nx.shortest_path(G, source=1, target=30)
print("Đường đi ngắn nhất từ 1 đến 30:")
print(path)
```

```python
all_paths = dict(nx.all_pairs_shortest_path(G))
print("All shortest paths (một phần):")
print({k: all_paths[k] for k in list(all_paths)[:3]})
```

```python
clustering = nx.clustering(G)
print("Clustering của từng node:")
print(clustering)

print("Average clustering:", nx.average_clustering(G))
```

```python
if nx.is_connected(G):
    print("Diameter:", nx.diameter(G))
    print("Radius:", nx.radius(G))
else:
    print("Graph không liên thông nên không tính được diameter")
```
